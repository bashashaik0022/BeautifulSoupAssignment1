Q2 : Write a python program to display IMDBs Top rated 100 movies data (i.e. Name, IMDB rating, Year of 
release) and save it in form of a CSV file.

```python
Q2_pages = np.arange(1, 100, 50)
for page in Q2_pages:
    url = "https://www.imdb.com/search/title/?groups=top_100&start=" + str(page)
    results = requests.get(url)
    soup = BeautifulSoup(results.text, "html.parser")
    movie_div = soup.find_all("div", class_="lister-item mode-advanced")
    for movieSection in movie_div:
        name = movieSection.h3.a.text
        titles.append(name)
        ratings = movieSection.strong.text
        imdb_ratings.append(ratings)
        year = movieSection.h3.find("span", class_="lister-item-year").text
        years.append(year)        
        
IMBD_Movies_List = pd.DataFrame({"Movie": titles,"Year of Release": years,"IMBD Rating": imdb_ratings})
```

```python
df = pd.DataFrame.drop_duplicates(IMBD_Movies_List)
```

```python
Q2 = df.sort_values(by="IMBD Rating",ascending=False)[0:100]
Q2
```

```python
#Saving File To CSV File
Q2.to_csv('TopIMBDList.csv')
```

Q3: Write a python program to display IMDBs Top rated 100 Indian movies data (i.e. Name, IMDB rating, Year 
of release) and save it in form of a CSV file. 

```python
Q3_pages = np.arange(1, 100, 50)
for page in Q3_pages:
    url = "https://www.imdb.com/india/top-rated-indian-movies/?sort=ir,desc&mode=simple&page=1"
    Q3_results = requests.get(url)
    Q3_soup = BeautifulSoup(Q3_results.text, "html.parser")
    movie_div1 = Q3_soup.find_all("div", class_="lister-item mode-advanced")
    for Q3_movieSection in movie_div1:
        name = Q3_movieSection.h3.a.text
        titles.append(name)
        ratings = Q3_movieSection.strong.text
        imdb_ratings.append(ratings)
        year = Q3_movieSection.h3.find("span", class_="lister-item-year").text
        years.append(year)        
        
IMBD_Indian_Movies_List = pd.DataFrame({"Movie": titles,"Year of Release": years,"IMBD Rating": imdb_ratings})
```

```python
df1 = pd.DataFrame.drop_duplicates(IMBD_Indian_Movies_List)
```

```python
Q3 = df1.sort_values(by="IMBD Rating",ascending=False)
Q3
```

```python
#Saving File To CSV File
IMBD_Indian_Movies_List.to_csv('TopIndianIMBDList.csv')
```

8  Write a python program to extract information about the local weather from the National Weather Service 
website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day 
extended forecast display for the city. The data should include period, short description, temperature and 
description. 

```python
pages = np.arange(1, 100, 50)
for page in pages:
    url = "https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YFexQtris2w" + str(page)
    results = requests.get(url)
results
```

```python
soup = BeautifulSoup(results.text, "html.parser")
soup
```

```python
Days = soup.find_all('div',class_="col-sm-2 forecast-label")
Days
```

```python
day_frequency = []
for i in Days:
    day_frequency.append(i.get_text().replace('\n',''))
day_frequency
```

```python
Description = soup.find_all('div',class_="col-sm-10 forecast-text")
Description
```

```python
day_description = []
for i in Description:
    day_description.append(i.get_text().replace('\n',''))
day_description
```

```python
Weather_Details = pd.DataFrame({"Day Wise Data": day_frequency,"Description": day_description})
Weather_Details
```
